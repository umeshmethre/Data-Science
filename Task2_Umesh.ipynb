```python
def area_of_circle(radius):
    return 3.14 * radius * radius

r = float(input("Enter radius: "))
print("Area of circle:", area_of_circle(r))
```

```python
def largest_of_three(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

print("Largest number:", largest_of_three(10, 25, 15))
```

```python
def string_length(s):
    count = 0
    for _ in s:
        count += 1
    return count

text = input("Enter a string: ")
print("Length of string:", string_length(text))
```

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

n = int(input("Enter a number: "))
print("Prime number" if is_prime(n) else "Not a prime number")
```

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

n = int(input("Enter n: "))
print("Fibonacci number:", fibonacci(n))
```

```python
def reverse_string(s):
    if s == "":
        return s
    else:
        return reverse_string(s[1:]) + s[0]

text = input("Enter a string: ")
print("Reversed string:", reverse_string(text))
```

```python
n = int(input("Enter a number: "))
sum = 0

for i in range(1, n + 1):
    sum += i

print("Sum =", sum)
```

```python
numbers = [10, 21, 4, 45, 66, 93, 8]
count = 0

for num in numbers:
    if num % 2 == 0:
        count += 1

print("Number of even numbers:", count)
```

```python
for num in range(2, 101):
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, end=" ")
```
