```python
import numpy as np
```

```python
#Write a NumPy program to find the sum of all elements in an array [10, 20, 30]

arr = np.array([10, 20, 30])
print(np.sum(arr))
```

```python
#Write a NumPy program to find the shape of the array [[1, 2, 3], [4, 5, 6]].

arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr.shape)
```

```python
#Write a NumPy program to find the maximum element in [5, 12, 3, 9].

arr = np.array([5, 12, 3, 9])
print(np.max(arr))
```

```python
#Write a NumPy program to add two arrays [1, 2, 3] and [4, 5, 6].

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
print(np.add(arr1, arr2))
```

```python
#Write a NumPy program to check the data type of an array [1, 2, 3].

arr = np.array([1, 2, 3])
print(arr.dtype)
```
